```python
def replace_titles(x):
    title = x['Title']
    if title in ['Capt','Col','Major', 'Dr', 'Rev']:
        return 'Officer'
    elif title in [ "Jonkheer",'Don','Sir','the Countess', 'Dona']:
        return 'Royalty'
    elif title in ['Mme', 'Mlle','Miss']:
        return 'Miss'
    elif title in ['Ms','Mr','Mrs']:
        return 'Mrs'
    elif title=='Master':
            return 'Master'
    else:
        return title
titanic['Title'] = titanic.apply(replace_titles, axis=1)

titanic
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

x1= titanic[['Sex']] 
y1= titanic['Survived']

from sklearn.model_selection import train_test_split

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.20, random_state=1)


print("number of test samples :", x1_test.shape[0])
print("number of training samples:",x1_train.shape[0])
```

According to the previous analysis, sex seems to be the most important feature that affects the survival rate. That's why, we will use it in the simple logistic regrssion.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1_train = sc.fit_transform(x1_train)
X1_test = sc.transform(x1_test)
```

```python
logreg1 = LogisticRegression()   #build our logistic model
logreg1.fit(X1_train, y1_train)  #fitting training data
y1_pred  = logreg1.predict(X1_test)    #testing model’s performance

print("Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
print("Precision:",metrics.precision_score(y1_test, y1_pred))
print("Recall:",metrics.recall_score(y1_test, y1_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y1_test,y1_pred))
```

```python
confusion_matrix = pd.crosstab(y1_test,y1_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

TP=92,FP=18,FN=18,TN=51

Well, the classification rate is almost 80%, it is considered as good accuracy. The presicion (positive predictive value) gives an idea about how much the positive predictions of the model are correct. When this logistic regression model predicted passengers are going to survive, this prediction is correct and the passenger really survived,74% of the time. Recall (sensivity) determines what proportion of actual positives (test) was identified correctly. This model has a recall of 0.74, it correctively identifies 74% of the survivors. Accuracy works best if false positives and false negatives have similar cost (This is our case). If the cost of false positives and false negatives are very different, it’s better to work with F1 score which is the weighted average of Precision and Recall.

```python
y1_pred_proba = logreg1.predict_proba(X1_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y1_test,  y1_pred_proba)

auc = metrics.roc_auc_score(y1_test, y1_pred_proba)
plt.plot(fpr,tpr,label="Survived, auc="+str(auc))

plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')

plt.legend(loc=4)
plt.show()

plt.show()
```

The Receiver Operator Characteristic (ROC) curve is used to separate the signal (True positive rate) from the noise (false positive rate). It is clear that we have higher values of TP and TN. The area under the curve (AUC) is the measure of the ability of a classifier to distinguish between classes and is used as a summary of the ROC curve. In our case 0.5<auc = 0.78<1.Our classifier is able to detect more numbers of TP and TN than FN and FP. This is a good classifier. 

This simple logistic regrssion is efficient in predicting the survival of passengers based on their Sex. Nevertheless, there is many features that affect the survival of passengers other than sex. Let's use a model based on the most important features to predict if a passenger will survive or not. 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

x = titanic[['Sex','Age','Pclass','Fare','Embarked','Title','Alone','Family_Size']] 
y = titanic['Survived']


from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)


print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
model1 = LogisticRegression()   #build our logistic model
model1.fit(X_train, y_train)  #fitting training data
y_pred  = model1.predict(X_test)    #testing model’s performance
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix = pd.crosstab(y_test,y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

TP=92,FP=17,FN=18,TN=49

The classification rate is almost 80%, it is considered as good accuracy. The precision score of the this model is 0.75. When it predicted passengers are going to survive, this prediction was correct and the passenger really survived,75% of the time. The model has a recall of 0.74, it correctively identifies 74% of the survivors. The number of FP is slightly different of FN. In this case, F1 score (74%) could be interesting in our analysis. 

```python
y_pred_proba = model1.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)

auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="Survived, auc="+str(auc))

plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')

plt.legend(loc=4)
plt.show()

plt.show()
```

It is clear that we have higher values of TP and TN. 0.5<auc = 0.86<1.Our classifier is able to detect more numbers of TP and TN than FN and FP. This is a good classifier.

The feature selection allowed us to choose the 4 most important features ['Sex', 'Pclass','Title', 'Alone'] to predict if a passenger will survive or not. The accuracy increased from 0.80 to 0.8. The precision increased from 0.75 to 0.81. It means that this model is more efficient in predicting true positives. When it predicts a passenger is going to survive, it is correct 81% of the time. The Recall decreased from 0.73 to 0.67 because we have more TP (from 93 to 99) and more FN (from 18 to 23). 

This model allowed us to have more TP and less TN. In contrast, we have more FN and less TN. Well, in our case it is important to have a better prediction of TP. Less TN means that there is less passengers that were predicted to survive but they didn't. We think that if the aim of this model is to enhance safety to save more passengers in the case of boat accident, then it is important to not overestimate people who survived.

The aim of SMOTE method is to addressing imbalanced datasets is to oversample the minority class. In our case it enhanced the recall of the model but the accuracy and precision decreased. Thus, the best way to standardize the features is the use of Standard Scaler.

Regularization generally refers the concept that there should be a complexity penalty for more extreme parameters. The idea is that just looking at the training data and not paying attention to how extreme one's parameters are leads to overfitting. A high value of C tells the model to give high weight to the training data, and a lower weight to the complexity penalty. A low value tells the model to give more weight to this complexity penalty at the expense of fitting to the training data. Basically, a high C means "Trust this training data a lot", while a low value says "This data may not be fully representative of the real world data, so if it's telling you to make a parameter really large, don't listen to it".
